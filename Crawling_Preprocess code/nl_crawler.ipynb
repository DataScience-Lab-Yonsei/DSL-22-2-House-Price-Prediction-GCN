```python
url = "https://m.land.naver.com/cluster/ajax/articleList"
HEAD = {
        'User-Agent': "PostmanRuntime/7.20.0",
        'Accept': "*/*",
        'Cache-Control': "no-cache",
        'Postman-Token': "adbba748-cb85-4fb4-8f6a-4be441f19cc3",
        'Host': "m.land.naver.com",
        'Accept-Encoding': "gzip, deflate",
        'Connection': "keep-alive",
        'cache-control': "no-cache"
        }
```

```python
def get_soup(DONG,LAT,LON,PAGE,거래방법, z):
##행정동 및 위도 경도##변수 입력 
    dong={"cortarNo":DONG}
    lat={"lat":LAT}
    lon={"lon":LON}
    page={"page":PAGE}
  ##고정값
    con1= {"rletTpCd":"DDDGG%3AGM"} ## 검색물건
    con2= {"tradTpCd":거래방법} ##거래방법 A1 : 매매, B1 전세 , B2 월세, B3 단기임대
    con3={"showR0":""}    ## 그외조건 없음
    con4 = {"z":7}  ## 축척
  ##검색조건
    space1={"spcMin":"0"} 
    space2={"spcMax":"900000000"} ## 무한
    price1={"dprcMin":"0"} 
    price2={"dprcMax":"10000000"}
  ##mapping 고정값
    rgt = {"rgt": str(float(lat["lat"])+0.3475542)}
    lft = {"lft": str(float(lat["lat"])-0.3475542)}
    top = {"top": str(float(lon["lon"])+0.0538928)}
    btm = {"btm": str(float(lon["lon"])-0.0538928)}
    qstring={**con1,**con2,**con3,**con4,**space1,**space2,**price1,**price2,**dong,**lat,**lon,**lft,**rgt,**top,**btm,**page}
    result=requests.request("GET",url,headers=HEAD,params=qstring,timeout=5)
    text=json.loads(result.text)
    return text
```

```python
z = 7
k=0
last_pg =0
i = 0
for r in range(1, 6) :
    k = r*100  
    for GU in code:  
        datas=pd.DataFrame()

        DONG=str(GU["DONG"]*100000)
        LAT=str(GU["LAT"])
        LON=str(GU["LON"])
        GU_ENG = GU['gu_eng']

        save_file= f'{GU_ENG}{r}.csv'
        
        l = 1
        while l <= 100:
            i = k + l
            try:
                time.sleep(2)
                soup = get_soup(DONG,LAT,LON,i,'A1:B1:B2:B3',str(z))
                
                if soup["more"]==True:
                    data=soup["body"]
                    datas=datas.append(data)

                    if l% 10 ==0: print(GU_ENG,': ',i) 
                    l +=1    
                elif soup["more"]==False:
                    data_fin = datas[colname]
                    data_fin.to_csv(save_file)
                    print("no page")
                    last_pg = i
                    break
                                
            except:
                print('차단당함 6분 휴식')
                print(f'{GU_ENG}구 {i}까지 완료')
                last_pg = i
                data_fin = datas[colname]
                data_fin.to_csv(save_file)
                time.sleep(360)
                continue       
            
        data_fin = datas[colname]
        data_fin.to_csv(save_file)
    
    print(f'모든 구 {last_pg}까지 완료')
```
